```python
import pandas as pd
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
df = pd.merge(ratings, movies, on="movieId")
df_final = df[['userId', 'movieId', 'rating', 'title', 'genres']]
df_final.to_csv("final_dataset.csv", index=False)
df=pd.read_csv("final_dataset.csv")
print(df.head().to_string(index=False))
```

```python
user_movie_matrix = df.pivot_table(
    index='userId',
    columns='title',
    values='rating'
).fillna(0)
user_movie_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_similarity = cosine_similarity(user_movie_matrix)

user_similarity_df = pd.DataFrame(
    user_similarity,
    index=user_movie_matrix.index,
    columns=user_movie_matrix.index
)
user_similarity_df.head()
```

```python

avg_rating = df.groupby('title')['rating'].mean()
rating_count = df.groupby('title')['rating'].count()
top_movies = pd.DataFrame({
    'avg_rating': avg_rating,
    'rating_count': rating_count
})
top_movies = top_movies[top_movies['rating_count'] > 50]
top_movies = top_movies.sort_values(by='avg_rating', ascending=False)
def get_top_movies_for_user(user_id, n=10):
    user_movies = user_movie_matrix.loc[user_id]
    watched_movies = user_movies[user_movies > 0].index
    filtered_movies = [
        movie for movie in top_movies.index
        if movie not in watched_movies
    ]
    return filtered_movies[:n]
```

```python
def recommend_by_genre(movie_name, n=5):
    
    # Get genres of selected movie
    genre = df[df['title'] == movie_name]['genres'].values
    
    if len(genre) == 0:
        return []
    
    genre_list = set(genre[0].split('|'))  # convert to set

    # Function to count matching genres
    def match_count(x):
        movie_genres = set(x.split('|'))
        return len(genre_list.intersection(movie_genres))

    # Apply matching
    df['match_score'] = df['genres'].apply(match_count)

    # Keep only strong matches (>=2 genres common)
    similar_movies = df[df['match_score'] >= 2]

    # Remove same movie
    similar_movies = similar_movies[similar_movies['title'] != movie_name]

    # Sort by match score + rating
    similar_movies = similar_movies.sort_values(
        by=['match_score', 'rating'],
        ascending=False
    )

    return similar_movies['title'].drop_duplicates().head(n).tolist()
```

```python
def recommend_movies(user_id, n=5):

    similar_users = user_similarity_df[user_id].sort_values(ascending=False)[1:6]

    user_movies = user_movie_matrix.loc[user_id]
    watched_movies = user_movies[user_movies > 0].index

    recommendations = {}

    for sim_user, similarity_score in similar_users.items():
        sim_user_movies = user_movie_matrix.loc[sim_user]

        for movie, rating in sim_user_movies.items():
            if movie not in watched_movies and rating > 0:
                if movie not in recommendations:
                    recommendations[movie] = similarity_score * rating
                else:
                    recommendations[movie] += similarity_score * rating

    sorted_movies = sorted(
        recommendations.items(),
        key=lambda x: x[1],
        reverse=True
    )

    return [movie[0] for movie in sorted_movies[:n]]

```

```python
def full_recommendation(user_id):

    # Get one movie user watched
    user_movies = user_movie_matrix.loc[user_id]
    watched = user_movies[user_movies > 0].index
    
    if len(watched) == 0:
        return
    
    movie_name = watched[0]  # pick first movie

    print("🔥 Top Picks for You:")
    print(get_top_movies_for_user(user_id))

    print("\n🎯 Because you watched:", movie_name)
    print(recommend_by_genre(movie_name))

    print("\n👥 Users like you also watched:")
    print(recommend_movies(user_id))
```
